```python
#edad = ['E5_14','E15_24','E25_34','	E35_54','E55_74','E75+']
fig, ax = plt.subplots(figsize=(20,10),sharey=True)

p1 = ax.bar(df_prom['Pais'],df_prom['Promedio de suicidios/100k'],label="General",color="#7ba5ed")
p2 = ax.bar(df_prom['Pais'],df_prom['Prom suicidios masculinos /100k'], label="Hombres",color="#1a8510",alpha=0.5)

#Permite girar las etiquetas
plt.xticks(rotation=90,fontsize='large')
plt.legend(loc="best",fontsize='x-large')
plt.show()

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(20,10))
plt.xticks(rotation=90,fontsize='large')
plt.legend(loc="best",fontsize='x-large')

p1 = ax.bar(df_prom['Pais'],df_prom['E5_14'],label = "5-14",zorder=6)
p2 = ax.bar(df_prom['Pais'],df_prom['E15_24'],label = "15-24",zorder=5)
p3 = ax.bar(df_prom['Pais'],df_prom['E25_34'],label = "25-34",zorder=4)
p4 = ax.bar(df_prom['Pais'],df_prom['E35_54'],label = "35-54",zorder=3)
p5 = ax.bar(df_prom['Pais'],df_prom['E55_74'],label = "55-74",zorder=2)
p6 = ax.bar(df_prom['Pais'],df_prom['E75+'],label = "75+",zorder=1)

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(20,10))
plt.xticks(rotation=90,fontsize='large')
plt.legend(loc="best",fontsize='x-large')

p6 = ax.bar(df_prom['Pais'],df_prom['E75+'],label = "75+ años",zorder=1)
p5 = ax.bar(df_prom['Pais'],df_prom['E55_74'],label = "55-74 años",zorder=2)
p4 = ax.bar(df_prom['Pais'],df_prom['E35_54'],label = "35-54 años",zorder=3)
p3 = ax.bar(df_prom['Pais'],df_prom['E25_34'],label = "25-34 años",zorder=4)
p2 = ax.bar(df_prom['Pais'],df_prom['E15_24'],label = "15-24 años",zorder=5)
p1 = ax.bar(df_prom['Pais'],df_prom['E5_14'],label = "5-14 años",zorder=6)

p6mean = df_prom['E75+'].mean()
p5mean = df_prom['E55_74'].mean()
p4mean = df_prom['E35_54'].mean()
p3mean = df_prom['E25_34'].mean()
p2mean = df_prom['E15_24'].mean()
p1mean = df_prom['E5_14'].mean()

plt.axhline(p6mean, color='b', linestyle='-')
plt.axhline(p5mean, color='xkcd:orange', linestyle='-')
plt.axhline(p4mean, color='g', linestyle='-')
plt.axhline(p3mean, color='xkcd:red', linestyle='-')
plt.axhline(p2mean, color='xkcd:purple', linestyle='-')
plt.axhline(p1mean, color='xkcd:brown', linestyle='-')

plt.legend(loc="best")

plt.title("Promedio de la densidad de suicidios/100k por edad",fontsize='xx-large')

plt.show()
```
